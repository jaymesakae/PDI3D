## Derivadas no dominio espacial

### Informações sobre uma função f(x) no ponto x = x0

f(x0): valor da função no ponto x0.

$\frac{df(x)}{dx}$ |x = x0: deriva primeira, ou taxa de crescimento ou redução da função

$\frac{d²f(x)}{dx²}$ |x = x0 : derivada segunda, ou a demonstração da concavidade da função (negativa ou positiva)

### Aproximação de uma função por série de Taylor

a aproximação de ordem 0, o valor de uma função próxima de x, somado ao delta x, é igual ao valor de f(xo)

f(x0 + $\Delta x$) = f(x0)

e de ordem 1, a função é declarada como:

f(x0 + $\Delta x$) = f(x0) + $\frac {df}{dx}$|x=x0 * $\Delta x$ 

a aproximação de ordem 2:

f(x0 + $\Delta x$) = f(x0) + $\frac {df}{dx}$|x=x0 * $\Delta x$ + $\frac {1}{2}$ * $\frac {d²f}{fx²}$|x=x0 * $\Delta x²$

## Em 2D ficaria:

$\frac {d²f(x, y)}{dx²}$|x=x0, y=y0 : derivada segunda em x

$\frac {d²f(x, y)}{fy²}$|x=x0, y=y0 : derivada segunda em y

$\frac {d²f(x, y}{dxdy}$|x=x0, y=y0 : derivada cruzada

$\frac {df(x, y)}{dy}$ ~ f(x, y+1) - f(x, y)

$\frac {d}{dx}$ * ($\frac {df(x, y)}{dy}$) = $\frac {d²f(x, y)}{dx*dy}$ ≃ f(x+1, y+1) - f(x, y+1) - (f(x+1, y) - f(x, y))

ou melhor dizendo:

$\frac {d²f(x, y)}{dx * dy}$ ≃ 1*f(x, y) - 1*f(x, y+1) - 1*f(x+1, y) + 1*f(x+1, y+1)

### Matriz Hessiana

Basicamente, armazena a matriz H, que contem na diagonal segundaria a derivada segunda de x e de y, respectivamente, e na diagonal principal tera a derivada cruzada

ou seja:

H = 

[$\frac {d²f}{dx²}$  $\frac {d²f}{dx * dy}$]

[$\frac {d²f}{dy * dx}$   $\frac {d²f}{dy²}$]]

Em 3D, ficará:

obs: chamaremos a derivada segunda com a repetição da variavel pra facilitar a visualização, mas eh o caralho ai em cima

[fxx   fxy   fxz]

[fxy   fyy   fyz]

[fzx   fzy   fzz]

V = vetor da matematica (n sei fazer isso em markdown ainda)

* Vr0: posição do ponto

* $\Delta Vr$ : intervalo em torno do ponto Vr0

f(Vr0 + $\Delta Vr$) = f(Vr0) + $\left. \frac {df}{dVr} \rigth|{Vr = Vr0}{T}$

OU, 

f(Vr0 + $\Delta Vr$) = f(Vr0) + $\Delta F^T$ * $\Delta Vr$ + $\Delta Vt^T$ * H * $\Delta Vr$

sendo:

* H = Matriz Hessiana

* $\Delta Vf^T$ = Gradiente

Usando esse contexto da matriz hessiana, temos que considerar os conceitos de auto-vetores e auto-valores, que basicamente, mudaremos a base de referencia para que fiquei consideravelmente, apenas com a diagonal segundaria de valores da matrix, ou seja

[Ipp Ipr Ipc] -> [$\lambda 1$ 0 0]

[Icp Irr Irc] -> [0 $\lambda 2$ 0]

[Icp Icr Icc] -> [0 0 $\lambda 3$]

Os auto-vetores tendem a se alinhar em um angulo de 90°, dado uma estrutura tubular, indicando entao a maior variação dos angulos.

Se a matriz hessiana for pequena, todas as variações de intensidade sobre a imagem é baixa, enquanto se um for alto e outro for baixo, ele forma uma linha sobre a direção do auto-vetor alto, se os auto-valores forem altos, então a imagem de formação dele será alta, ou seja, como uma bola de intensidade.

Uma forma de ficar mais facil de analisar visualmente, é interessante fazer uma suavisação sobre a imagem, ou uma convolução sobre a visinhança do pixel que esta sendo analisado.

Os auto-valores e auto-vetores ja são implementados em funções prontas como scypy da vida ai

2D: \  \$\lambda1 \  \ \lambda2$
3D: $\lambda1 \lambda2 \lambda3$

~ 0 ~ 0 ~ 0 ~ 0 ~ 0 -> região uniforma

~0 >>0 ~0 >>0 >>0 -> estrutura tubular escura

~0 <<0 ~0 <<0 <<0 -> estrutura tubular brilhante

\>>0 >>0 >>0 >>0 >>0 -> blob(ou bolha) ou uma forma parecida com uma bolha escura

\<<0 <<0 <<0 <<0 <<0 -> bolha brilhante

------ ------ ~0 ~0 >>0 -> prato (ou plate) escuro-

------ ------ ~0 ~0 <<0 -> prato brilhante

### Tarefa: fazer a matriz hessiana no ponto r

```python
def hessian(img_in, r):
    plan, row, col = np.array(r) + 1
    print(plan, row, col)
    # derivada do plano, linha e coluna
    
    img = np.pad(img_in, pad_width = 1, mode = 'edge')
    
    der_plan = -2*img[plan, row, col] + img[plan-1 , row, col] + img[plan+1, row, col]
    der_row = -2*img[plan, row, col] + img[plan, row - 1, col] + img[plan, row+1, col]
    der_col = -2*img[plan, row, col] + img[plan, row, col - 1] + img[plan, row, col+1]
    
    # derivada cruzada
    der_cross_rc = img[plan, row, col] - img[plan, row+1, col] - img[plan, row, col+1] + img[plan, row+1, col+1]
    der_cross_pr = img[plan, row, col] - img[plan+1, row, col] - img[plan, row+1, col] + img[plan+1, row+1, col]
    der_cross_pc = img[plan, row, col] - img[plan, row, col+1] - img[plan+1, row, col] + img[plan+1, row, col+1]
    
    H = np.zeros((3, 3))
    
    H[0,0] = der_plan
    H[1, 1] = der_row
    H[2, 2] = der_col
    
    H[0, 1] = der_cross_pr
    H[1, 0] = der_cross_pr
    
    H[0, 2] = der_cross_pc
    H[2, 0] = der_cross_pc
    
    H[1, 2] = der_cross_rc
    H[2, 1] = der_cross_rc
    
    return H

def hessian_all(img):
    filter_der = np.array([1, -2, 1])
    filter_der_cruz = np.array([[1, -1],
                                [-1, 1]])
    filter_der_p = filter_der.reshape(3, 1, 1)
    filter_der_r = filter_der.reshape(1, 3, 1)
    filter_der_c = filter_der.reshape(1, 1, 3)
    
    filter_der_pr = filter_der_cruz.reshape(2, 2, 1)
    filter_der_pc = filter_der_cruz.reshape(2, 1, 2)
    filter_der_rc = filter_der_cruz.reshape(1, 2, 2)
    
    img_pad = np.pad(img, pad_width=1, mode='edge')
    Ipp = convolve(img_pad, filter_der_p, mode = 'same')
    Irr = convolve(img_pad, filter_der_r, mode = 'same')
    Icc = convolve(img_pad, filter_der_c, mode = 'same')
    
    
    Ipr = convolve(img_pad, filter_der_pr, mode = 'same')
    Ipc = convolve(img_pad, filter_der_pc, mode = 'same')
    Irc = convolve(img_pad, filter_der_rc, mode = 'same')
    
    H_all = np.array(([[Ipp, Ipr, Ipc],
                       [Ipr, Irr, Irc],
                       [Ipc, Irc, Icc]]))
    H_all_unpad = H_all[:,:,1:-1, 1:-1, 1:-1]
    H_all_unpad = np.transpose(H_all_unpad, (2, 3, 4, 0, 1))
    
    return H_all_unpad
```

```python
#eigen vem do ingles pra auto valores
def hessian2d_eigenvalues(img):
    """
    Calcula oas autovalores da Hessiana para uma imagem 2D
    """
    
    num_rows, num_cols = img.shape
    img = img[None]
    
    H_all = hessian_all(img)
    
    eigenvalues = np.zeros((num_rows, num_cols, 2))
    
    for row in range(num_rows):
        for col in range(num_cols):
            # Indice 0 porque adicionamos uma dimensão
            H = H_all[0, row, col]
            
            # Descarta as derivadas ao longo do plano
            H = H[1:, 1:] 
            
            eigenvalues[row, col] = np.linalg.eigvalsh(H)
    return eigenvalues

def sort_eigenvalues(eigenvalues):
    num_rows, num_cols = eigenvalues.shape[0:2]
    for row in range(num_rows):
        for col in range(num_cols):
            l1, l2 = eigenvalues[row, col]
            if abs(l1)>abs(l2):
                eigenvalues[row, col] = (l2, l1)
            
    return eigenvalues

def tubeness_2d(img, eig_ratio, signal_strength, sigma):
    """filtro Frangi para detectar estruturas tubulares, essa função detecta as estruturas escuras"""
    
    img_smooth = ndi.gaussian_filter(img.astype(float), sigma)
    
    eigenvalues = hessian2d_eigenvalues(img_smooth)
    eigenvalues_s = sort_eigenvalues(eigenvalues)
    
    eigval_ratio2 = 2*eig_ratio**2
    signal_strength2 = 2*signal_strength**2
    l1 = eigenvalues[:,:, 0]
    l2 = eigenvalues[:,:, 1]
    
    Rb = l1/(l2+1e-10)
    Rb2 = Rb**2
    S2 = l1**2 + l2**2

    # tub_val é o V da formula vista na teoria
    tub_val = np.exp(-Rb2/eigval_ratio2) * (1-np.exp(-S2/signal_strength2))
    ind = np.nonzero(l2<0)
    tub_val[ind] = 0
    
    return tub_val
```
