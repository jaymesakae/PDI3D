## Transformation of a set of points

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from numpy import cos,sin

# Array containing four poins. Each point is a column of the array.
points = np.array([[0., 5., 5., 0],
                   [0., 0., 3., 3]])

plt.plot(points[0], points[1], 'o')
```

```python
def add_ones(points):
    '''Add an additional coordinate having value 1 to each point. This
       is called a homogeneous coordinate.'''

    num_points = points.shape[1]
    points_h = np.concatenate((points, np.ones((1, num_points))), axis=0)
        
    return points_h

def transform(points, T):
    '''Apply transformation T to the set of points. It is assumed that
       points is represented in homogeneous coordinates'''
        
    return T@points # Matrix multiplication, same as np.dot(T, points)
```

### Translation

```python
def get_translation_matrix(tx, ty):
    Tt = np.array([[1., 0., tx],
                   [0., 1., ty],
                   [0., 0., 1.]])
    return Tt

tx = 5
ty = 10
Tt = get_translation_matrix(tx, ty)
# Applying the transformation
points_h = add_ones(points)
points_t = transform(points_h, Tt)

plt.plot(points[0], points[1], 'o')
plt.plot(points_t[0], points_t[1], 'o')
plt.legend(('Original', 'Transformed'))
```

```python
def get_rotation_matrix(angle):
    
    T = np.array([[cos(angle), -sin(angle), 0.],
                  [sin(angle), cos(angle), 0.],
                  [0., 0., 1.]])
    return T

angle = np.pi/10.
Tr = get_rotation_matrix(angle)
points_r = transform(points_h, Tr)

plt.plot(points[0], points[1], 'o')
plt.plot(points_r[0], points_r[1], 'o', ms=3)
```

```python
def get_affine_matrix():
    # Essa matrix usada, eh uma matrix generica, então esses valores mudam ao decorrer da utilização
    
    T = np.array([[1., 1.3, 1.2],
                  [0.5, 0.3, 2.0],
                  [0., 0., 1.]])
    return T

Ta = get_affine_matrix()
points_a = transform(points_h, Ta)

plt.plot(points[0], points[1], 'o')
plt.plot(points_a[0], points_a[1], 'o')
```

Caso quisermos fazer a imagem voltar ao original quando necessario, é fazer a operação com a matrix inversa

```python
Ta_inv = np.linalg.inv(Ta)
points_rec = transform(points_a, Ta_inv)

plt.plot(points[0], points[1], 'o')
plt.plot(points_rec[0], points_rec[1], 'o')
```

## Image geometrical transformations

Temos duas abordagem que podem ser usadas para a transformação geometrica de imagens Ii (de input); e uma imagem de saida Io (image output)

### 1° Abordagem

Para cada pixel r-> de Ii, calculamos um r-> = T * r-> e associamos Io(r->) = Ii(r->)

Para evitar problemas como, dado dois pixeis da imagem e fazermos a operação de trasnformação, podem existir burados, o modo que podemos tratar esse problema é o mapeamento direto, onde ele fará uma interporlação dos valores para evitar resse problema

### 2° Abordagem (ou mapeamentoinverso)

Para cada pixel r'-> de Io, calculamos r-> = $T^{-1}$ * r'-> e associamos Io(r'->) = Ii(r->)

```python
def get_similarity_matrix(tr, tc, angle, s):
    T = np.array([[s*cos(angle), -s*sin(angle), tr],
                  [s*sin(angle), s*cos(angle), tc],
                  [0, 0, 1]])
    return T

def apply_similarity_transform(point, T, reference_point):
    
    point -= reference_point
    
    point_h = np.array([point[0], point[1], 1]).transpose()
    point_transform = T@point_h
    
    point_transform = point_transform[0:2, 0]
    
    point_transform += reference_point
    
    return point_transform

def apply_similarity_transform_arr(points, T, reference_point):
    """Considerando que cada linha de points possui um ponto"""
    num_points = points.shape[0]
    points -= reference_point
    
    points_h = np.ones((num_points, 3))
    points_h[:, :2] = points
    
    points_transf = T@points_h.transpose()
    points_transf = points_transf[:2].transpose()
    points_transf += reference_point
    
    
    return points_transf
    
def bilinear_interpol(img, point):
    '''Return the intensity at the position indicated by 'point'. The position can be
       any floating point value inside the range [0, img.shape[0]-1) for the row and
       [0, img.shape[1]-1) for the column.'''
       
    row_int = int(point[0])
    col_int = int(point[1])
    row_frac = point[0]-row_int
    col_frac = point[1]-col_int
    I00 = img[row_int, col_int]
    I01 = img[row_int, col_int+1]
    I10 = img[row_int+1, col_int]
    I11 = img[row_int+1, col_int+1]
    val = (I00*(1-row_frac)*(1-col_frac) + I01*(1-row_frac)*col_frac + 
           I10*row_frac*(1-col_frac) + I11*row_frac*col_frac)
    
    return val

def transform_img(img, T, reference_point, out_img_corners):
    """O out_img_corners: [ponto esquerdo superior, ponto direito inferior], ou seja, a extremidade da nova imamge
    Para evitar perda de dados da imagem que faremos as operações, por exemplo, em uma rotação
    perderiamos as informações das extremidades da imagem
    """
    
    out_img_upper_left = out_img_corners[0].astype(int)
    out_img_lower_right = out_img_corners[1].astype(int)
    
    img_width = out_img_lower_right[1] - out_img_upper_left[1]
    img_height = out_img_lower_right[0] - out_img_upper_left[0]
    T_inv = np.linalg.inv(T)
    
    out_img = np.zeros((img_width, img_height))
    
    for row in range(img_height):
        row_coord = row+out_img_upper_left[0]
        for col in range(img_width):
            col_coord = col+out_img_upper_left[1]
            point = np.array([row_coord, col_coord])
            row_coord_in, col_coord_in = apply_similarity_transform(point, T_inv, reference_point)
            
            if row_coord_in >= 0 and row_coord_in < img.shape[0] \
            and col_coord_in >= 0 and col_coord_in < img.shape[1]:
                val = bilinear_interpol(img, (row_coord_in, col_coord_in))
                out_img[row, col] = img[row_coord_in, col_coord_in]
                
    return out_img

img = plt.imread('../cameraman.tiff')
Ts = get_similarity_matrix(20, 5, np.pi/30, 1.)
reference_point = np.array([0, 0])
out_img_corners = np.array([[0., 0.], [255., 255.]])
out_img = transform_img(img, Ts, reference_point, out_img_corners)
```

```python
points = np.array([[0., 5.], 
                   [5., 0.],
                   [0., 0.],
                   [3., 3.]])
T = get_similarity_matrix(20, 5, np.pi/10, 2.)
ref_point = np.array([4., 7.])
points_t = apply_similarity_transform_arr(points, T, ref_point)

plt.plot(points[0], points[1], 'o')
plt.plot(points_t[0], points_t[1], 'o')
```
