## Image geometrical transformations

We will see how to apply geometric transformation to an image

```python


import numpy as np
import matplotlib.pyplot as plt
from numpy import cos,sin

```

```python
def get_similarity_transform_matrix(tr, tc, angle, s):
    '''Return a matrix describing the similarity transformation having translation
       (tr, tc), rotation given by angle and scaling given by s.'''
    
    Ts = np.array([[s*cos(angle), -s*sin(angle), tr],
                   [s*sin(angle),  s*cos(angle), tc],
                   [          0.,            0., 1.]])
    return Ts

def apply_similarity_transform(point, T, reference_point):
    '''Apply similarity transformaiton to a point represented by the variable 
       'point'. 'T' is the transformaiton matrix and 'reference_poin't sets the anchor
       point for the rotation and scaling.'''

    point = np.array(point)
    # Subtract reference_point
    reference_point = np.array(reference_point)
    point -= reference_point
    # Add value 1 to point
    point_h = np.array([[point[0], point[1], 1]]).transpose()
    
    # Transform
    point_transf = T@point_h
    # Recover original coordinates (remove ones from homogeneous coordinates)
    point_transf = point_transf[:2, 0]
    # Add reference point back
    point_transf += reference_point
    
    return point_transf

def apply_similarity_transform_arr(points, T, reference_point):
    '''Apply similarity transformaiton to a list of points represented by the 
       variable point. T is the transformaiton matrix and reference_point sets 
       the anchor point for the rotation and scaling.'''    
    
    points = np.array(points)
    reference_point = np.array(reference_point)
    num_points = points.shape[0]
    points -= reference_point
    points = points.T
    points_h = np.ones((3, num_points))
    points_h[:2] = points
    
    points_transf = T@points_h
    points_transf = points_transf[:2].T
    points_transf += reference_point
    
    return points_transf

def transform_image(img, T, reference_point, out_img_corners):
    '''Transforms an image using transformaiton matrix T. 'reference_point' 
       sets the anchor point for the rotation and scaling and 'out_img_corners'
       sets the bounds for the output image.'''
    
    out_img_upper_left = out_img_corners[0]
    out_img_bottom_right = out_img_corners[1]
    
    img_width = out_img_bottom_right[1] - out_img_upper_left[1]
    img_height = out_img_bottom_right[0] - out_img_upper_left[0]
    T_inv = np.linalg.inv(T)

    out_img = np.zeros((img_height, img_width))
    for row in range(img_height):
        new_row_coord = row + out_img_upper_left[0]
        for col in range(img_width):
            new_col_coord = col + out_img_upper_left[1]
            orig_row_coord, orig_col_coord = apply_similarity_transform(
                       (new_row_coord, new_col_coord), T_inv, reference_point)

            if orig_row_coord>=0 and orig_row_coord<img.shape[0]-1 \
                    and orig_col_coord>=0 and orig_col_coord<img.shape[1]-1:
                orig_row_coord = int(round(orig_row_coord))
                orig_col_coord = int(round(orig_col_coord))
                out_img[row, col] = img[orig_row_coord, orig_col_coord]
                
    return out_img

def transform_image_interpol(img, T, reference_point, out_img_corners):
    '''Transforms an image using transformaiton matrix M. If the coordinates in the 
       original image do not fall exactly at a pixel position, the intensity is 
       inteprolated. 'reference_point' sets the anchor point for the rotation and 
       scaling and 'out_img_corners' sets the bounds for the output image.'''
    
    out_img_upper_left = out_img_corners[0]
    out_img_bottom_right = out_img_corners[1]
    
    img_pad = np.pad(img, pad_width=(0, 1), mode='reflect')
    
    img_width = out_img_bottom_right[1] - out_img_upper_left[1]
    img_height = out_img_bottom_right[0] - out_img_upper_left[0]
    T_inv = np.linalg.inv(T)

    out_img = np.zeros((img_height, img_width))-1
    for row in range(img_height):
        row_coord = row + out_img_upper_left[0]
        for col in range(img_width):
            col_coord = col + out_img_upper_left[1]
            orig_row_coord, orig_col_coord = apply_similarity_transform(
                                (row_coord, col_coord), T_inv, reference_point)

            if orig_row_coord>=0 and orig_row_coord<img.shape[0] \
                          and orig_col_coord>=0 and orig_col_coord<img.shape[1]:
                val = bilinear_interpol(img_pad, (orig_row_coord,orig_col_coord))
                out_img[row, col] = val
                
    return out_img

def bilinear_interpol(img, point):
    '''Return the intensity at the position indicated by 'point'. The position can be
       any floating point value inside the range [0, img.shape[0]-1) for the row and
       [0, img.shape[1]-1) for the column.'''
       
    row_int = int(point[0])
    col_int = int(point[1])
    row_frac = point[0]-row_int
    col_frac = point[1]-col_int
    I00 = img[row_int, col_int]
    I01 = img[row_int, col_int+1]
    I10 = img[row_int+1, col_int]
    I11 = img[row_int+1, col_int+1]
    val = (I00*(1-row_frac)*(1-col_frac) + I01*(1-row_frac)*col_frac + 
           I10*row_frac*(1-col_frac) + I11*row_frac*col_frac)
    return val
    
```

```python


img = plt.imread('../cameraman.tiff')
plt.figure(figsize=[8,8])
plt.imshow(img, 'gray')

```

```python
# Define where to crop the output image
out_img_upper_left = (0, 0)
out_img_bottom_right = (255, 255)

Ts = get_similarity_transform_matrix(20, 5, np.pi/30, 1)
reference_point = (img.shape[0]//2, img.shape[1]//2)

out_img_corners = [out_img_upper_left, out_img_bottom_right]
out_img = transform_image(img, Ts, reference_point, out_img_corners)

plt.figure(figsize=[8,8])
plt.imshow(out_img, 'gray')
```

```python
# Transform image using linear interpolation
out_img = transform_image_interpol(img, Ts, reference_point, 
                                   out_img_corners)

plt.figure(figsize=[8,8])
plt.imshow(out_img, 'gray')
```

# Using python libraries

```python
import scipy.ndimage as ndi
from skimage import transform

def get_similarity_transform_matrix_ref(t_row, t_col, angle, s, ref_point):
    
    Tt = get_similarity_transform_matrix(-ref_point[0], -ref_point[1], 0, 1)
    Ts = get_similarity_transform_matrix(t_row, t_col, angle, s)
    Tt_inv = get_similarity_transform_matrix(ref_point[0], ref_point[1], 0, 1)
    
    Ts_final = Tt_inv@Ts@Tt
    
    return Ts_final

Ts = get_similarity_transform_matrix_ref(20, 5, np.pi/30, 1, reference_point)
```

```python


#Scipy
T_inv = np.linalg.inv(Ts)
out_img = ndi.affine_transform(img, T_inv, order=1)
plt.figure(figsize=[8,8])
plt.imshow(out_img, 'gray')

```

```python


#Scikit-image

# Scikit-image uses the xy convention. Thus, we need to swap translation values and use -angle. 
Ts = get_similarity_transform_matrix_ref(5, 20, -np.pi/30, 1, reference_point)
Ts_sk = transform.AffineTransform(Ts)
# We can also create the transformation using specific values for the parameters. Unfortunately, 
# the function does not let us set the reference point.
#Ts_sk = transform.AffineTransform(translation=(5, 20), rotation=-np.pi/30, scale=1, shear=0)

Ts_sk_inv = Ts_sk.inverse

print(Ts)
print(Ts_sk)

```

```python


out_img = transform.warp(img, Ts_sk_inv, order=1)
plt.figure(figsize=[8,8])
plt.imshow(out_img, 'gray')

```

# Perspective transform

```python


def get_perspective_transform_matrix(v1, v2, img_shape):

    Tp = np.array([[1,  0,    0],
                   [0,  1,    0],
                   [v1, v2, 1]])
    
    T1 = get_similarity_transform_matrix_ref(0., -img_shape[0]//2, 0, 1, (0,0))
    T2 = get_similarity_transform_matrix_ref(0., img_shape[0]//2, 0, 1, (0,0))

    Tp = T2@Tp@T1
    
    return Tp

Tp = get_perspective_transform_matrix(0.001, 0., img.shape)
T_inv = np.linalg.inv(Tp+1e-10)   # +1e-10 is necessary for conditioning the inverse
out_img = transform.warp(img, T_inv, order=1)
plt.figure(figsize=[8,8])
plt.imshow(out_img, 'gray')

```
